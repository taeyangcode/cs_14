```python
# Question 1


class Employee:
    def __init__(self, name, number):
        self.__name = name
        self.__number = number

    def get_name(self):
        return self.__name

    def get_number(self):
        return self.__number

    def set_name(self, name):
        self.__name = name

    def set_number(self, number):
        self.__number = number


class ProductionWorker(Employee):
    def __init__(self, name, number, shift_number, hourly_pay_rate):
        super().__init__(name, number)
        self.__shift_number = shift_number
        self.__hourly_pay_rate = hourly_pay_rate

    def get_shift_number(self):
        return self.__shift_number

    def get_hourly_pay_rate(self):
        return self.__hourly_pay_rate

    def set_shift_number(self, shift_number):
        self.__shift_number = shift_number

    def set_hourly_pay_rate(self, hourly_pay_rate):
        self.__hourly_pay_rate = hourly_pay_rate


def main():
    name = input("Employee Name:")
    employee_number = int(input("Employee Number:"))
    shift_number = int(input("Employee Shift Number:"))
    hourly_pay_rate = float(input("Hourly Pay Rate:"))

    employee = ProductionWorker(name, employee_number, shift_number, hourly_pay_rate)

    print(f"Employee Name: {employee.get_name()}")
    print(f"Employee Number: {employee.get_number()}")
    print(f"Employee Shift Number: {employee.get_shift_number()}")
    print(f"Employee Hourly Pay Rate: ${format(employee.get_hourly_pay_rate(), '.2f')}")


main()
```

```python
# Question 2


class ShiftSupervisor(Employee):
    def __init__(self, name, number, annual_salary, annual_production_bonus):
        super().__init__(name, number)
        self.__annual_salary = annual_salary
        self.__annual_production_bonus = annual_production_bonus

    def get_name(self):
        return "This employee's name is classified."

    def get_annual_salary(self):
        return self.__annual_salary

    def get_annual_production_bonus(self):
        return self.__annual_production_bonus

    def set_annual_salary(self, annual_salary):
        self.__annual_salary = annual_salary

    def set_annual_production_bonus(self, annual_production_bonus):
        self.__annual_production_bonus = annual_production_bonus


def main():
    name = input("Employee Name:")
    employee_number = int(input("Employee Number:"))
    annual_salary = float(input("Employee Annual Salary:"))
    annual_production_bonus = float(input("Employee Annual Production Bonus:"))

    employee = ShiftSupervisor(
        name, employee_number, annual_salary, annual_production_bonus
    )

    print(f"Employee Name: {employee.get_name()}")
    print(f"Employee Number: {employee.get_number()}")
    print(f"Employee Annual Salary: ${format(employee.get_annual_salary(), '.2f')}")
    print(
        f"Employee Annual Production Bonus: ${format(employee.get_annual_production_bonus(), '.2f')}"
    )


main()
```

```python
import turtle


class Snowman:
    def __init__(self):
        self.screen = turtle.Screen()
        self.screen.title("Snowman Drawing")
        self.t = turtle.Turtle()
        self.t.speed(5)

    def draw_circle(self, x, y, radius):
        self.t.penup()
        self.t.goto(x, y - radius)
        self.t.pendown()
        self.t.circle(radius)

    def draw_snowman(self):
        # Draw bottom section
        self.draw_circle(0, -150, 60)
        # Draw midsection
        self.draw_circle(0, -50, 45)
        # Draw head
        self.draw_circle(0, 40, 30)

    def draw_hat(self):
        self.t.penup()
        self.t.goto(-20, 70)
        self.t.pendown()
        self.t.setheading(0)
        self.t.color("black")
        self.t.begin_fill()
        for _ in range(2):
            self.t.forward(40)
            self.t.right(90)
            self.t.forward(10)
            self.t.right(90)
        self.t.end_fill()

        self.t.penup()
        self.t.goto(-15, 80)
        self.t.pendown()
        self.t.begin_fill()
        for _ in range(2):
            self.t.forward(30)
            self.t.right(90)
            self.t.forward(20)
            self.t.right(90)
        self.t.end_fill()

    def draw_nose(self):
        self.t.penup()
        self.t.goto(0, 30)
        self.t.pendown()
        self.t.color("orange")
        self.t.begin_fill()
        self.t.circle(5)
        self.t.end_fill()

    def draw_eyes(self):
        self.t.color("black")
        # Draw left eye
        self.t.penup()
        self.t.goto(-10, 45)
        self.t.pendown()
        self.t.begin_fill()
        self.t.circle(3)
        self.t.end_fill()

        # Draw right eye
        self.t.penup()
        self.t.goto(10, 45)
        self.t.pendown()
        self.t.begin_fill()
        self.t.circle(3)
        self.t.end_fill()


class BetterSnowman(Snowman):
    def __init__(self):
        super().__init__()
        self.t.speed(0)

    def draw_hat(self):
        self.t.penup()
        self.t.goto(-20, 70)
        self.t.pendown()
        self.t.setheading(0)
        self.t.color("black")
        self.t.begin_fill()
        for _ in range(2):
            self.t.forward(40)
            self.t.right(90)
            self.t.forward(10)
            self.t.right(90)
        self.t.end_fill()

        self.t.penup()
        self.t.goto(-15, 70)
        self.t.pendown()
        self.t.begin_fill()
        for _ in range(2):
            self.t.forward(30)
            self.t.left(90)
            self.t.forward(100)
            self.t.left(90)
        self.t.end_fill()

    def draw_eyes(self):
        self.t.color("black")
        # Draw left eye
        self.t.penup()
        self.t.goto(-10, 45)
        self.t.pendown()
        self.t.begin_fill()
        self.t.circle(3)
        self.t.end_fill()

        # Draw right eye
        self.t.penup()
        self.t.goto(10, 45)
        self.t.pendown()
        self.t.begin_fill()
        self.t.setheading(0)
        for _ in range(2):
            self.t.forward(10)
            self.t.left(90)
            self.t.forward(5)
            self.t.left(90)
        self.t.end_fill()

    def draw_nose(self):
        self.t.penup()
        self.t.goto(0, 30)
        self.t.pendown()
        self.t.color("orange")
        self.t.begin_fill()
        self.t.goto(5, 30)
        self.t.goto(0, 15)
        self.t.goto(-5, 30)
        self.t.goto(5, 30)
        self.t.end_fill()


def main():
    # Will draw a snowman

    # Create Snowman object that builds a snowman
    snow1 = BetterSnowman()
    snow1.draw_snowman()
    snow1.draw_hat()
    snow1.draw_eyes()
    snow1.draw_nose()
    snow1.t.hideturtle()
    snow1.screen.mainloop()


main()
```
